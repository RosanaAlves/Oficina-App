```python
df['De qual Cidade você é?'].unique()
```

```python
# Remover espaços no início e no final de cada string:
df['De qual Cidade você é?'] = df['De qual Cidade você é?'].str.strip()
```

```python
contagem = df['De qual Cidade você é?'].value_counts().sort_values()

ax = contagem.plot(
    kind='barh',
    figsize=(10, 8),
    color=['#FF9AA2', '#FFB7B2', '#FFDAC1'],
    edgecolor='black'
)

# Títulos e formatação
ax.set_title('Distribuição de Idades', fontsize=14)
ax.set_xlabel('Total de Respondentes')
ax.set_ylabel('')
ax.grid(axis='x', linestyle=':')

# Adicionar rótulos de dados
for barra in ax.patches:
    ax.text(
        barra.get_width() + 1,
        barra.get_y() + barra.get_height()/2,
        f'{int(barra.get_width())}',
        ha='left',
        va='center'
    )

plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Dados de exemplo
contagem = df['De qual Cidade você é?'].value_counts().reset_index()
contagem.columns = ['De qual Cidade você é?', 'Total']

# Plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Total', y='De qual Cidade você é?', data=contagem, palette='viridis')

# Ajusta o eixo X para inteiros
plt.xticks(range(0, int(max(contagem['Total'])) + 1, step=1))  # Passo de 1
plt.xlabel('Contagem (Inteiros)')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Exemplo: Gráfico de barras com contagem de cidades
contagem = df['De qual Cidade você é?'].value_counts()
contagem.plot(kind='barh', figsize=(10, 10))

# Ajusta o eixo X para mostrar apenas inteiros
plt.xticks(range(0, int(max(contagem)) + 1, 1))  # Passo de 1 em 1
plt.xlabel('Número de Ocorrências (Inteiros)')
plt.show()
```

# Contagem de idades

```python
contagem_idades = df.groupby('Qual a sua idade?').size()
print(contagem_idades)
```

```python
plt.figure(figsize=(8, 6))
import matplotlib.pyplot as plt

# Dados
contagem = df['Qual a sua idade?'].value_counts()

# Plot
plt.figure(figsize=(8, 6))
plt.pie(contagem, labels=None, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Distribuição de Idades (%)')
plt.axis('equal')  # Gráfico redondo
plt.legend(contagem.index, loc='upper right')
plt.show()
```

# Perfil do motorista

```python
plt.figure(figsize=(8, 6))
import matplotlib.pyplot as plt

# Dados
contagem = df['Qual seu perfil como motorista? '].value_counts()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(contagem, labels=None, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Perfil do motorista (%)',pad=16)
plt.axis('equal')  # Gráfico redondo
plt.legend(contagem.index, loc='upper right', bbox_to_anchor=(1.5, 1))
plt.show()
```

### 'Quantos km você roda por mês, em média?'

```python
plt.figure(figsize=(8, 6))
import matplotlib.pyplot as plt

# Dados
contagem = df['Quantos km você roda por mês, em média?'].value_counts()

# Plot
plt.figure(figsize=(8, 8))
plt.pie(contagem, labels=None, autopct='%1.1f%%', startangle=90, colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Média da quantidade de km que se roda por mês',pad=16)
plt.axis('equal')  # Gráfico redondo
plt.legend(contagem.index, loc='upper right', bbox_to_anchor=(1.5, 1))
plt.show()
```
