### O que você mais prioriza em uma oficina?

```python
# Converter tudo para string e tratar NaN/nulos
import pandas as pd
from collections import Counter
respostas = df['O que você mais prioriza em uma oficina? (Marque até 2) ']

respostas_limpas = []
for item in respostas:
    if pd.isna(item):  # Se for NaN ou None
        continue  # Ignora ou substitui por string vazia ("")
    respostas_limpas.append(str(item))  # Converte para string

print(respostas_limpas)
```

```python
# Separar todas as opções individuais
todas_opcoes = []
for resposta in respostas_limpas:
    opcoes = [opcao.strip() for opcao in resposta.split(",")]
    todas_opcoes.extend(opcoes)

# Contar frequências
contagem = Counter(todas_opcoes)

# Converter para DataFrame
df_contagem = pd.DataFrame.from_dict(contagem, orient='index').reset_index()
df_contagem.columns = ['Opção', 'Contagem']
df_contagem = df_contagem.sort_values(by='Contagem', ascending=False)

print(df_contagem)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 8))
bars = plt.barh(df_contagem['Opção'], df_contagem['Contagem'], color='skyblue')
plt.title('O que você mais prioriza em uma oficina? (Marque até 2) ')
plt.xticks(rotation=45, ha='right')  # Melhor legibilidade
plt.bar_label(bars)  # Adiciona contagem em cima das barras
plt.tight_layout()
plt.show()
```
